```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from category_encoders import HashingEncoder
```

### Loading the Train and test datasets after Feature Extraction

```python
train = pd.read_csv("../data/train_features.csv")
test = pd.read_csv("../data/test_features.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
label_encoders = {}

# Loop through all columns and apply Label Encoding to object (categorical) columns
for col in train.columns:
    if train[col].dtype == 'object':
        le = LabelEncoder()
        
        # Combine train and test for consistent encoding
        combined_data = pd.concat([train[col], test[col]], axis=0)
        le.fit(combined_data)

        # Encode
        train[col] = le.transform(train[col])
        test[col] = le.transform(test[col])
        
        # Store encoder for future use (e.g., decoding)
        label_encoders[col] = le
```

```python
# View results
print("Encoded train data sample:")
train.head()

print("\nEncoded test data sample:")
test.head()
```

### Save encoded datasets

```python
# Save encoded datasets
train.to_csv("../data/train_encoded.csv", index=False)
test.to_csv("../data/test_encoded.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

## Create a DataFrame manually with your results

```python
comparison_df = pd.DataFrame({
    'Model': ['AdaBoost', 'XGBoost'],
    'Accuracy': [0.84, 0.84],
    'Precision (Class 1)': [0.83, 0.87],
    'Recall (Class 1)': [0.36, 0.36],
    'F1-Score (Class 1)': [0.50, 0.51],
})

print("Model Comparison Table:")
comparison_df
```

## Bar plot for each metric

```python
metrics = ['Accuracy', 'Precision (Class 1)', 'Recall (Class 1)', 'F1-Score (Class 1)']

plt.figure(figsize=(12, 6))
for i, metric in enumerate(metrics, 1):
    plt.subplot(2, 2, i)
    sns.barplot(x='Model', y=metric, data=comparison_df, palette='viridis')
    plt.title(metric)
    plt.ylim(0, 1)
    plt.ylabel('')
    plt.xlabel('')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
